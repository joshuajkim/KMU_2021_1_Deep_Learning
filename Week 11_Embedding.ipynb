# IMDB 데이터를 사용한 Embedding 학습

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
MaxFeatures = 10000
MaxLen = 20
```

```python
(train_input, train_target), (test_input, test_target) = imdb.load_data(num_words=MaxFeatures)

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
print(train_input[0])
print(train_target[0])
```

* Padding

```python
train_input = pad_sequences(train_input, maxlen=MaxLen)
test_input = pad_sequences(test_input, maxlen=MaxLen)

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
print(train_input[0])
print(train_target[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Embedding
```

```python
model = Sequential()

# max_features + 1을 한 이유: Padding으로 인해 0이 생겼기 때문임!

model.add(Embedding(
    input_dim=MaxFeatures + 1,
    output_dim=8,
    input_length=MaxLen
))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['acc']
)
```

```python
history = model.fit(
    train_input,
    train_target,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import pandas as pd
```

```python
MaxFeatures = 2000

tokenizer = Tokenizer(num_words=MaxFeatures)
tokenizer.fit_on_texts(input_list)

input_list = tokenizer.texts_to_sequences(input_list)
print(len(input_list))
```

```python
for idx, (input, target) in enumerate(zip(input_list, target_list)):
    if idx == 5:
        break
    print(input)
    print(target)
```

```python
tokenizer.word_index
```

```python
word2idx_dict = {k:v for k, v in tokenizer.word_index.items()}
idx2word_dict = {v:k for k, v in tokenizer.word_index.items()}
```

```python
MaxLen = 40

data_input = pad_sequences(input_list, maxlen=MaxLen)
print(data_input.shape)
```

```python
for idx in range(5):
    print(data_input[idx])
```

* 지금 label_list가 1과 0으로 구분되어 있는데, batch로 model fitting할 때를 대비하여 서로 섞어줘야 한다.
* 아래의 코드로 shuffle을 한다.

```python
index_arr = np.arange(data_input.shape[0])

np.random.shuffle(index_arr)

data_input = data_input[index_arr]

data_target = np.asarray(target_list)
data_target = data_target[index_arr]

print(data_input.shape, data_target.shape)
```

* 사전에 미리 학습된 GloVe Embedding 가져오기

```python
word2coef_dict = {}

f = open('/content/drive/MyDrive/03. Kookmin AI Big Data MBA/Semester 3_032021-062021/2. Deep Learning/Jupyter Notebook/data_Week 11/glove.6B.50d.txt',
         encoding='utf8')

for line in f:

    value_list = line.split()

    word = value_list[0]
    coef_arr = np.asarray(value_list[1:], dtype='float32')

    word2coef_dict[word] = coef_arr

f.close()
```

```python
word2coef_dict.get('coffee')
```

```python
word2coef_dict.get('korea')
```

* word2idx와 Word2Coef_dict를 서로 연결시켜주자!

```python
dim_cnt = 50

word2coef_arr = np.zeros((MaxFeatures, dim_cnt))

for word, idx in word2idx_dict.items():
    if idx < MaxFeatures:
        coef_arr = word2coef_dict.get(word)
        if coef_arr is not None:
            word2coef_arr[idx] = coef_arr
```

```python
print(word2coef_arr.shape)
print(word2coef_arr[100])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense

model = Sequential()

model.add(Embedding(
    input_dim=MaxFeatures,
    output_dim=dim_cnt,
    input_length=MaxLen
))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

* Model에 GloVe Embedding Load하기

```python
model.layers[0].set_weights([word2coef_arr])
model.layers[0].trainable = False

model.summary()
```

```python
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['acc']
)
```

```python
history = model.fit(
    data_input,
    data_target,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```
